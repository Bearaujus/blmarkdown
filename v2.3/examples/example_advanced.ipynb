### Importing Library

```python
from blmarkdown import Learning, LearningData 
```

### Variable Instantiation

```python
learning_subject = 'PHP Basic'
```

```python
author = 'Haryo Bagas Assyafah'
```

```python
footer_data = {'Linkedin' : '[https://www.linkedin.com/in/bearaujus/](https://www.linkedin.com/in/bearaujus/)',
              'Main Github Page' : '[https://github.com/bearaujus/](https://github.com/bearaujus/)'}
```

### Implementing blmarkdown

```python
# Learning Instantiation
learning = Learning(learning_subject, author, footer_data)
```

```python
# Data1
data1 = LearningData('PHP Syntax', 'https://github.com/bearaujus/learning/tree/main/php/1_basic_php/l01')
data1.add('Standard Output')
data1.add('Variable')
data1.add('Operator', ['Concat', 'Assigment', 'Comparison', 'Identity', 'Logic'])

# Adding Data1 to Learning
learning.add(data1)
```

```python
# Preview Data1 Markdown
data1.markdown()
```

```python
# Data2
data2 = LearningData('Control Structure', 'https://github.com/bearaujus/learning/tree/main/php/1_basic_php/l02')
data2.add('Loop', ['For', 'While', 'Do While', 'For Each'])
data2.add('Branching', ['If', 'Else If', 'Else'])

# Adding Data2 to Learning
learning.add(data2)
```

```python
# Preview Data2 Markdown
data2.markdown()
```

```python
# This function will open a new tab in your default browser for viewing the generated markdown
learning.preview()
```

### Get Learning Markdown

```python
learning.markdown()
```

### Save Learning Markdown

```python
learning.save_markdown()
```
